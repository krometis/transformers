# Activation Functions
This notebook just plots commonly-used activations. There's not much else to see here.

```python
import torch, torch.nn as nn
import matplotlib.pyplot as plt
```

```python
act_funcs = [ nn.ReLU(), nn.Sigmoid(), nn.Tanh(), nn.ELU(), nn.Softplus(), nn.GELU() ]
act_lbls  = [ "ReLU", "Sigmoid", "Tanh", "ELU", "Softplus", "GELU" ]

fig, ax = plt.subplots(1,1,figsize=(5,5))
x = torch.linspace(-5, 5, 100)
for func,lbl in zip(act_funcs,act_lbls):
    ax.plot(x.numpy(),func(x).numpy(),label=lbl)
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')
fig.legend(bbox_to_anchor=(0.15, 0.95), loc='upper left')
fig.tight_layout()
fig.savefig('figures/activations.png')
```
